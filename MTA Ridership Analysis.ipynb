### Data Preparation

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.head()
```

```python
df['Subways: Total Estimated Ridership'].groupby(df['Date'].dt.month).sum()
```

```python
df_2021 = df[(df['Date'] >= '2021-01-01') &
            (df['Date'] < '2022-01-01')]
df_2021
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
```

```python
monthly_ridership_2021 = df_2021['Subways: Total Estimated Ridership'].groupby(df['Date'].dt.month).sum()
monthly_ridership_2021 = pd.DataFrame([months, monthly_ridership_2021]).T
monthly_ridership_2021.rename(columns = {0: 'Month', 1: 'Total Subway Ridership 2021'}, inplace = True)
monthly_ridership_2021
```

```python
monthly_ridership_2021.info()
```

```python
monthly_ridership_2021['Total Subway Ridership'] = monthly_ridership_2021['Total Subway Ridership'].astype(int)
monthly_ridership_2021.info()
```

```python
df_2022 = df[(df['Date'] >= '2022-01-01') &
            (df['Date'] < '2023-01-01')]
df_2022
```

```python
monthly_ridership_2022 = df_2022['Subways: Total Estimated Ridership'].groupby(df['Date'].dt.month).sum()
monthly_ridership_2022 = pd.DataFrame([months, monthly_ridership_2022]).T
monthly_ridership_2022.rename(columns = {0: 'Month', 1: 'Total Subway Ridership 2022'}, inplace = True)
monthly_ridership_2022
```

```python
monthly_ridership_2022['Total Subway Ridership'] = monthly_ridership_2022['Total Subway Ridership'].astype(int)
monthly_ridership_2022.info()
```

```python
df_2023 = df[(df['Date'] >= '2023-01-01') &
            (df['Date'] < '2024-01-01')]
df_2023
```

```python
monthly_ridership_2023 = df_2023['Subways: Total Estimated Ridership'].groupby(df['Date'].dt.month).sum()
monthly_ridership_2023 = pd.DataFrame([months, monthly_ridership_2023]).T
monthly_ridership_2023.rename(columns = {0: 'Month', 1: 'Total Subway Ridership 2023'}, inplace = True)
monthly_ridership_2023
```

```python
monthly_ridership_2023['Total Subway Ridership'] = monthly_ridership_2023['Total Subway Ridership'].astype(int)
monthly_ridership_2023.info()
```

```python
df_2020 = df[(df['Date'] >= '2020-01-01') &
            (df['Date'] < '2021-01-01')]
df_2020
```

```python
monthly_ridership_2020 = df_2020['Subways: Total Estimated Ridership'].groupby(df['Date'].dt.month).sum()
monthly_ridership_2020 = pd.DataFrame([months, monthly_ridership_2020]).T
monthly_ridership_2020.rename(columns = {0: 'Month', 1: 'Total Subway Ridership'}, inplace = True)

monthly_ridership_2020 = monthly_ridership_2020['Total Subway Ridership'].fillna(value = int('0'))
monthly_ridership_2020 = pd.DataFrame([months, monthly_ridership_2020]).T
monthly_ridership_2020.rename(columns = {0: 'Month', 1: 'Total Subway Ridership 2020'}, inplace = True)
monthly_ridership_2020['Total Subway Ridership 2020'] = monthly_ridership_2020['Total Subway Ridership 2020'].astype(int)
monthly_ridership_2020.info()
```

```python
monthly_ridership_2020['Total Subway Ridership'] = monthly_ridership_2020['Total Subway Ridership'].astype(int, errors = 'ignore')
monthly_ridership_2020.info()
```

```python
#pd.merge(pd.merge(df1,df2,on='name'),df3,on='name')
monthly_ridership_per_year = pd.merge(pd.merge(pd.merge(monthly_ridership_2020, monthly_ridership_2021, on = 'Month'), monthly_ridership_2022, on = 'Month'), monthly_ridership_2023, on = 'Month')
monthly_ridership_per_year
```

```python
monthly_ridership_per_year.to_csv('MTA Monthly Ridership.csv')
```
